```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
# from matplotlib import pyplot as plt
```

```python
import numpy as np

time_steps=500
max_KW=500


def log_cost(capacity, a=250, b=100):
    return a + b*(np.log(capacity))  

capacities = np.linspace(0, max_KW, time_steps)
costs=[]
for i in range(0,time_steps):
    costs.append(log_cost(i))

print(costs)

steps=3
step_width=max_KW/steps
step_costs=[]
for i in range(steps):
    start = int(i * step_width)
    end = int((i + 1) * step_width)
    # Average the cost over the step range
    avg_cost = np.mean([log_cost(capacity) for capacity in range(start, end)])
    step_costs.extend([avg_cost] * (end - start))

# Plot the log function
plt.figure(figsize=(10, 6))
plt.step(capacities, step_costs, where='pre', color='red', linestyle='--', label='Step Approximations')
plt.plot(capacities, costs, label='Capital Cost', color='blue')
plt.xlabel('Installed Capacity (kW)')
plt.ylabel('Capital Cost')
plt.title('Logarithmic Capital Cost Function')
plt.grid(True)
plt.legend()
plt.show()
```
